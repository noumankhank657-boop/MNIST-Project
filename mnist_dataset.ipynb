#### Importing Libraries

```python
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras import layers
```

```python
import warnings
warnings.filterwarnings('ignore')
```

#### Loading the Dataset and spliting training and testing data

```python
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()
```

#### Scaling the Images Array for better learning

```python
train_images = train_images / 255.0
test_images = test_images / 255.0
```

#### Model Arhitecture

```python
model = Sequential([
    layers.Flatten(input_shape=(28, 28)),
    layers.Dense(256, activation='relu'),
    layers.Dense(128, activation='relu'),
    layers.Dense(10, activation='softmax')
])
```

#### Model Compilation

```python
model.compile(optimizer='rmsprop',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

#### Model Training on the training data

```python
model.fit(train_images, train_labels, epochs=10, batch_size=32)
```

#### Evaluating Model on test data

```python
model.evaluate(test_images, test_labels)
```

#### Model Summary

```python
model.summary()
```

#### Resizing and Coverting an Image to numpy array

```python
import numpy as np
from PIL import Image, ImageOps
img = Image.open("test_image_7.png").convert("L")
img = img.resize((28, 28))
img_array = np.array(img)
normalized_array = img_array / 255.0
model_test_image = normalized_array.reshape(1, 28, 28)
```

#### Giving image to the model to predict the image

```python
predictions = model.predict(model_test_image)

print("\n--- Model Predictions ---")
print(f"Predicted Number is: {np.argmax(predictions)}")
```

#### Saving the model

```python
model.save('model.keras')
```
